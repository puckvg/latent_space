```python
import numpy as np
import pandas as pd
import os
```

```python
def get_inchi(filename):
    with open(filename, "r") as f:
        lines = [line.rstrip() for line in f]
    # keep last line 
    line = lines[-1]
    inchi_a, inchi_b = line.split()
    inchi_a = inchi_a.split("=")[-1]
    inchi_b = inchi_b.split("=")[-1]

    return inchi_a, inchi_b
```

```python
def assert_match(data_dir):
    counter = 0
    for filename in sorted(os.listdir(data_dir)):
        filename = data_dir + filename
        a, b = get_inchi(filename)
        if a != b:
            print(a, b)
            counter += 1
    return counter  
```

```python
# want around 3000 stable molecules
# need to delete those indexed by "bad molecules"
data_dir = "../data/stable_qm9/"
assert_match(data_dir)
```
