```python
import numpy as np
import pandas as pd
from rdkit import Chem
import os 
```

```python
# need to canonicalise SMILES
def canonical_SMILES(SMILES):
    mol = Chem.MolFromSmiles(SMILES)
    try:
        canon_SMILES = Chem.MolToSmiles(mol, canonical=True)
    except: 
        canon_SMILES = "NoSMILES"
    return canon_SMILES
```

```python
def get_SMILES(filename):
    with open(filename, "r") as f:
        lines = [line.rstrip() for line in f]
    # keep second last line 
    line = lines[-2]
    smiles_a, smiles_b = line.split()
    return canonical_SMILES(smiles_a), canonical_SMILES(smiles_b)
```

```python
def assert_SMILES(data_dir):
    counter = 0
    for filename in sorted(os.listdir(data_dir)):
        filename = data_dir + filename
        smiles_a, smiles_b = get_SMILES(filename)
        if smiles_a != smiles_b:
            print(filename)
            counter += 1
    return counter  
```

```python
# want around 3000 stable molecules
# need to delete those indexed by "bad molecules"
data_dir = "../data/stable_qm9/"
assert_SMILES(data_dir)
```
